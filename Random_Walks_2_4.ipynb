```python
T=100 # total time
N=10000 # no. of random walkers
xt = np.zeros((T,N))
plt.figure()
for n in range(N):
    times = np.arange(T)
    # print('times = {}'.format(times))
    traj = [0]
    xt[0,n] = 0
    for t in range(1,T):
        step = np.random.choice([1,-1])
        x = traj[-1] + step
        traj.append(x)
        xt[t,n] = x
    if n<5:
        plt.plot(times,traj,'*-')
plt.xlabel('time')
plt.ylabel('position')
# print(xt)
```

```python
 # Calculate <x(t)>
meanx = np.mean(xt,1) #sum over the walkers
print(meanx)
plt.figure()
plt.plot(times, meanx)
```

```python
# Calculate <x**2(t)>
xt2 = np.square(xt)
meanx2 = np.mean(xt2,1)
print(meanx2)
plt.figure()
plt.plot(times, meanx2, '.')
plt.plot(times,times)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define Gaussian function
def gaussian(x, mu, sigma):
    return (1.0 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * ((x - mu) / sigma)**2)

# Select time steps (e.g., t=10 and t=90)
t1, t2 = 10, 90
xt_t1 = xt[t1, :]  # Positions at t=10
xt_t2 = xt[t2, :]  # Positions at t=90

# Create histograms
counts_t1, bin_edges_t1 = np.histogram(xt_t1, bins=30, density=True)
counts_t2, bin_edges_t2 = np.histogram(xt_t2, bins=30, density=True)
bin_centers_t1 = (bin_edges_t1[:-1] + bin_edges_t1[1:]) / 2
bin_centers_t2 = (bin_edges_t2[:-1] + bin_edges_t2[1:]) / 2

# Fit Gaussian to t=10 data
popt_t1, pcov_t1 = curve_fit(gaussian, bin_centers_t1, counts_t1, p0=[0, np.sqrt(t1)])
mu_t1, sigma_t1 = popt_t1

# Fit Gaussian to t=90 data
popt_t2, pcov_t2 = curve_fit(gaussian, bin_centers_t2, counts_t2, p0=[0, np.sqrt(t2)])
mu_t2, sigma_t2 = popt_t2

# Plot histograms and fitted Gaussians
plt.figure(figsize=(10, 6))
plt.hist(xt_t1, bins=30, density=True, alpha=0.5, label=f't={t1} (data)')
plt.hist(xt_t2, bins=30, density=True, alpha=0.5, label=f't={t2} (data)')
plt.plot(bin_centers_t1, gaussian(bin_centers_t1, mu_t1, sigma_t1), 'r-', label=f't={t1} (fit: ={mu_t1:.2f}, ={sigma_t1:.2f})')
plt.plot(bin_centers_t2, gaussian(bin_centers_t2, mu_t2, sigma_t2), 'b-', label=f't={t2} (fit: ={mu_t2:.2f}, ={sigma_t2:.2f})')
plt.xlabel('Position')
plt.ylabel('Probability Density')
plt.legend()
plt.title('Gaussian Fit to Walker Positions at Different Times')
plt.show()
```

```python
print(f"Fitted mean at t={t1}: {mu_t1:.4f} (expected: 0)")
print(f"Fitted variance at t={t1}: {sigma_t1**2:.4f} (expected: {t1})")
print(f"Fitted mean at t={t2}: {mu_t2:.4f} (expected: 0)")
print(f"Fitted variance at t={t2}: {sigma_t2**2:.4f} (expected: {t2})")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
N_walkers = 1000  # Number of walkers
T = 100           # Total time steps
bias = 0.6        # Probability of stepping right (must be > 0.5 for bias)

# Initialize positions (all walkers start at 0)
positions = np.zeros((T, N_walkers))

# Simulate biased random walk
for t in range(1, T):
    # Generate random steps: +1 (right) with probability 'bias', else -1 (left)
    steps = np.where(np.random.rand(N_walkers) < bias, 1, -1)
    positions[t, :] = positions[t-1, :] + steps

# Plot trajectories (first 10 walkers)
plt.figure(figsize=(10, 5))
for n in range(10):
    plt.plot(positions[:, n], alpha=0.7, linewidth=1)
plt.xlabel("Time Step")
plt.ylabel("Position")
plt.title(f"Trajectories of Biased Random Walk (Bias = {bias})")
plt.grid(True)
plt.show()

# Plot position distribution at t=50 and t=T-1
plt.figure(figsize=(10, 5))
plt.hist(positions[50, :], bins=30, alpha=0.7, density=True, label=f"t=50")
plt.hist(positions[-1, :], bins=30, alpha=0.7, density=True, label=f"t={T-1}")
plt.xlabel("Position")
plt.ylabel("Probability Density")
plt.title(f"Position Distribution (Bias = {bias})")
plt.legend()
plt.grid(True)
plt.show()

# Fit Gaussian to the final distribution
mu, sigma = norm.fit(positions[-1, :])
print(f"Final distribution:  = {mu:.2f},  = {sigma:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
m = 1000        # Number of walkers
T = 500         # Number of time steps
step_size = 1   # Size of each step

# Initialize all walkers at origin
x = np.zeros((m, T+1))
y = np.zeros((m, T+1))

# Random walk
for t in range(1, T+1):
    angles = 2 * np.pi * np.random.rand(m)  # Uniform direction
    dx = step_size * np.cos(angles)
    dy = step_size * np.sin(angles)
    x[:, t] = x[:, t-1] + dx
    y[:, t] = y[:, t-1] + dy

# Plotting trajectories of first 10 walkers
plt.figure(figsize=(10, 6))
for i in range(10):
    plt.plot(x[i], y[i], lw=1)
plt.title('Trajectories of 10 Random Walkers')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.axis('equal')
plt.show()

# Final position distribution
plt.figure(figsize=(8, 6))
plt.hist2d(x[:, -1], y[:, -1], bins=50, cmap='hot')
plt.colorbar(label='Number of Walkers')
plt.title('Final Position Distribution (2D Histogram)')
plt.xlabel('X')
plt.ylabel('Y')
plt.axis('equal')
plt.show()
```
